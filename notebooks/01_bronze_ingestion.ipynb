```python
# Bronze Layer - Raw data ingestion

df = spark.read.option("header", "true")\
               .option("inferSchema", "true")\
               .csv("Files/healthcare_dataset.csv")
```

```python
# Check data

print(f"Total rows: {df.count()}")
print(f"Total colums: {len(df.columns)}")
df.printSchema()
```

```python
# Rename columns - remove spaces

df_renamed = df.toDF(*[c.replace(" ", "_").replace("/", "_") for c in df.columns])

print(df_renamed.columns)
```

```python
# Save as Bronze Delta table

df_renamed.write.format("delta")\
        .mode("overwrite")\
        .saveAsTable("bronze_patient_records")

print("Bronze table created successfully!")
```
